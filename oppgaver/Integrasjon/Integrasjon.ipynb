Oppgave 10  Finn koeffisientene til et annengradspolynom som g√•r gjennom punktene (1, 2), (2, 3) og (3, 1).


```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[1, 1, 1],
              [4, 2, 1],
              [9, 3, 1]])

b = np.array([2, 3, 1])

x = np.linalg.solve(A, b)
a, b, c = x
print("L√∏sning:")
print("a =", a)
print("b =", b)
print("c =", c)

def f(x,a,b,c):
    return x**2*a + x*b + c

x = np.linspace(0, 3, 1000)
y = f(x,a,b,c)
plt.plot([1, 2, 3], [2, 3, 1], 'o')
plt.plot(x, y)
plt.show()
```

Oppgave 22 Hvor mye ladning har passert gjennom motstanden n√•r kondensatoren er utladet?

```python
import numpy as np 

t0 = 1 # Ampere
R = 1 # Ohm
C = 1 # Farad

def strom(t):
    return t0 * np.exp(-t / (R * C))

t = np.linspace(0, int(1e6), int(1e7))
total_charge = 0
for i in range(len(t)):
    if strom(t[i]) < 1e-10:
        break
    total_charge += strom(t[i])

print(f"Total ladning gjennom motstanden: {total_charge:.2f} C")
```

Oppgave 31  Finn omkretsen av en ellipse med halvakser 1 og 2.
(Denne g√•r ikke med penn og papir. Pr√∏v b√•de riemannsummer og trapesmetoden.)

```python
import numpy as no

def elipse(t, a, b):
    return np.sqrt((a**2)*np.cos(t)**2 + (b**2)*np.sin(t)**2)

r1 = 2
r2 = 1
a = 0
b = np.pi/2
n = 100000
bredde = (b - a) / n
integral = 0
trapesum = 0

for i in range(n):
    x0 = a + i * bredde
    x1 = a + (i + 1) * bredde
    integral += elipse(x0, r1, r2) * bredde
    trapesum += (elipse(x0, r1, r2) + elipse(x1, r1, r2)) * bredde / 2

print(integral * 4)
print(trapesum * 4)
```
