Oppgave 3
Lag et script som tilnærmer løsningen til
y
′ = y
2 − y
3 hvor y(0) = 0.01,
ved trapesmetoden for t ∈ [0, 200].


```python
import numpy as np
import matplotlib.pyplot as plt

a = 0
b =200
N=20000
h=(b-a)/N
t=np.linspace(0,b,N+1)

x=np.zeros(N+1)
x[0]=0.01

def f(x):
    return x**2 - x**3

for i in range(N):
    x[i+1] = x[i] + h* f(x[i])
    x[i+1] = x[i] + (h/2)*(f(x[i]) + f(x[i+1]))

plt.plot(t,x)
plt.show()

print(x[-1])

```
