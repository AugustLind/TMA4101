5 Plott eksponensialfunksjonen og noen av taylorpolynomene i samme figur.


```python
import matplotlib.pyplot as plt
import numpy as np
import math as m

def f(x):
    return np.exp(x)

x = np.linspace(0, 5, 100000)
y = f(x)

def taylorPolynom(x,i):
    return sum([x**n/m.factorial(n) for n in range(0, i+1)])


plt.plot(x, y)
plt.plot(x, taylorPolynom(x, 2))
plt.plot(x, taylorPolynom(x, 3))
plt.plot(x, taylorPolynom(x, 6))
plt.plot(x, taylorPolynom(x, 10))
plt.legend(["$e^x$", "Taylorpolynom med grad 2", "Taylorpolynom med grad 3", "Taylorpolynom med grad 6", "Taylorpolynom med grad 10"])
plt.show()


```

6 Bruk taylorpolynomet til å approksimere 𝑒. Hvor mange ledd må du ha med for å få med alle
desimalene over?

```python
import matplotlib.pyplot as plt
import numpy as np
import math as m

def taylorPolynom(x,i):
    return sum([x**n/m.factorial(n) for n in range(0, i+1)])

gyldig = True
i = 1

while gyldig:
    e = taylorPolynom(1, i)
    if abs((e -  2.71828182845904523536)) < 1e-15:
        print("Taylorpolynom med grad", i, "er lik e")
        gyldig = False
    else:
        i += 1
```
